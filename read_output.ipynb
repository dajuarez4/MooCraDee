```python
import numpy as np
import pandas as pd
```

```python
lines = np.genfromtxt("./output.txt", dtype=str, delimiter="\n", skip_header=2)

def parse_line(s):
    s = str(s).strip()
    idx = int(s.split(":", 1)[0])

    p1, p2 = s.find("("), s.find(")")
    x_str, y_str = s[p1+1:p2].split(",")
    x, y = float(x_str), float(y_str)

    r1 = s.find("radii=") + len("radii=")
    r2 = s.find("px", r1)
    r = float(s[r1:r2].strip())

    sc1 = s.find("score=") + len("score=")
    score = float(s[sc1:].strip())

    return idx, x, y, r, score

df = pd.DataFrame(
    [parse_line(s) for s in lines if str(s).strip()],
    columns=["idx", "x_px", "y_px", "r_px", "score"]
)
```

```python
df = df.sort_values("r_px", ascending=False).reset_index(drop=True)
onepxkm=35.65/df["r_px"][0] 
df["r_px"].multiply(onepxkm)
```
